```python
from Bio.Seq import Seq
from Bio import SeqIO
```

# 1

```python
seqA = SeqIO.read('../Bio UAP/fasta/U61535.fasta', 'fasta')
seqB = SeqIO.read('../Bio UAP/fasta/U88957.fasta', 'fasta')  # ini kasi .seq biar gampang

print(f'Seq 1: {seqA.seq}')
print(f'Seq 2: {seqB.seq}')
```

```python
# 1
# 1A Frekuensi 
print(f'Count CGC in SeqA: {seqA.seq.count("CGG")}')
print(f'Count CGC in SeqB: {seqB.seq.count("CGG")}')

# 1B Len
print(f'\nSeq 1 Len: {len(seqA.seq)}')
print(f'Seq 2 Len: {len(seqB.seq)}')
print(f'Total Len: {len(seqA.seq) + len(seqB.seq)}')

# 1C Slicing
seqC = seqA.seq[-15:] + seqB.seq[:35]
print(f'\nSeq C: {seqC}')
print(f'Len Seq C: {len(seqC)}')
```

# 2

```python
# 2
import matplotlib.pyplot as plt
from collections import Counter

# 2A count freq A C G T in bar plot
counterA = Counter(seqA.seq)
counterB = Counter(seqB.seq)
counterC = Counter(seqC)

plt.bar(counterA.keys(), counterA.values())
plt.title('Freq Nucleotide Seq A')
plt.show()

plt.bar(counterB.keys(), counterB.values())
plt.title('Freq Nucleotide Seq B')
plt.show()

plt.bar(counterC.keys(), counterC.values())
plt.title('Freq Nucleotide Seq C')
plt.show()
```

```python
# 2B percentage
# Manual
def gc_manual(seq: Seq):
    gc = [i for i in seq if i == 'G' or i == 'C']
    return (len(gc)/len(seq)) * 100

def at_manual(seq: Seq):
    gc = [i for i in seq if i == 'A' or i == 'T']
    return (len(gc)/len(seq)) * 100

print(f'GC% Seq A: {gc_manual(seqA.seq)}')
print(f'GC% Seq B: {gc_manual(seqB.seq)}')
print(f'GC% Seq C: {gc_manual(seqC)}')

print(f'\nAT% Seq A: {at_manual(seqA.seq)}')
print(f'AT% Seq B: {at_manual(seqB.seq)}')
print(f'AT% Seq C: {at_manual(seqC)}')

# Built in
from Bio.SeqUtils import gc_fraction
print(f'\nGC% Seq A (built-in): {gc_fraction(seqA.seq) * 100}')
print(f'GC% Seq B (built-in): {gc_fraction(seqB.seq) * 100}')
print(f'GC% Seq C (built-in): {gc_fraction(seqC) * 100}')

print(f'\nAT% Seq A (built-in): {(1 - gc_fraction(seqA.seq)) * 100}')
print(f'AT% Seq B (built-in): {(1 - gc_fraction(seqB.seq)) * 100}')
print(f'AT% Seq C (built-in): {(1 - gc_fraction(seqC))  * 100}')
```

```python
# 2C Melting temp
from Bio.SeqUtils import MeltingTemp as mt

print(f'Tm Seq A: {mt.Tm_Wallace(seqA.seq)}')
print(f'Tm Seq B: {mt.Tm_Wallace(seqB.seq)}')
print(f'Tm Seq C: {mt.Tm_Wallace(seqC)}')
```

```python
# 2D Molecular weight
from Bio.SeqUtils import molecular_weight as mw
mol_a = mw(seqA.seq)
mol_b = mw(seqB.seq)
mol_c = mw(seqC)

print(f'\nMolecular Weight Seq A: {mol_a} Da')
print(f'Molecular Weight Seq B: {mol_b} Da')    
print(f'Molecular Weight Seq C: {mol_c} Da')
```

# 3

```python
# 3A Transcribe DNA seq A B C
mrnaA = seqA.seq.transcribe()
mrnaB = seqB.seq.transcribe()
mrnaC = seqC.transcribe()

print(f'\nmRNA Seq A: {mrnaA}')
print(f'mRNA Seq B: {mrnaB}')
print(f'mRNA Seq C: {mrnaC}')

# 3B Translate mRNA seq A B C
protA = mrnaA.translate()
protB = mrnaB.translate()
protC = mrnaC.translate()

print(f'\nProtein Seq A: {protA}')
print(f'Protein Seq B: {protB}')
print(f'Protein Seq C: {protC}')
```

# 4

```python
# 4A first 100 SeqA => seq D, find motif 'AUG' in seq E
seqD = seqA.seq[:100]
seqE = seqB.seq[:100]

print(f'\nSeq D (first 100 of Seq A): {seqD}')
print(f'Seq E (first 100 of Seq B): {seqE}')
```

```python
#4B Local lignment pairwise 2
from Bio import pairwise2
from Bio.pairwise2 import format_alignment
from Bio import Seq

seq_1 = Seq.Seq('GCATGTAGCTAGCATCAGCTAC')

alignmentD = pairwise2.align.localxx(seqD, seq_1)
for align in alignmentD:
    print(format_alignment(*align))
```

```python
# 4C
seq_2 = Seq.Seq('ATGCTAGTCGATCATTATTAGC')

alignmentE = pairwise2.align.localxx(seqE, seq_2)
for align in alignmentE:
    print(format_alignment(*align))
```

```python
# 4E levenshtein
from Levenshtein import distance 

lev_distance = distance(seqD, seqE)
print(f'\nLevenshtein Distance between Seq D and Seq E: {lev_distance}')
```
