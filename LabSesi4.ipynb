```python
# Alignment : bandingin 1 sequence denganyg lain utk liat similarity
from Bio.Seq import Seq
```

```python
dnaA = Seq("CGATCGAT")
dnaB = Seq("ACGTACGT")

# Alignment global: bandingin dna secara keseluruhan
from Bio import pairwise2
global_alignment = pairwise2.align.globalxx(dnaA, dnaB)  # xx = match score 1, mismatch 0
print("Global Alignment:")
for x in global_alignment:
    print(x) # ada banyak alognment karena ada banyak cara align lalu diambil yg tertinggi

# print format
from Bio.pairwise2 import format_alignment
print("\nFormatted Global Alignment:")
for x in global_alignment:
    print(format_alignment(*x))


# Alignment local: bandingin dna secara lokal / pakai subsequence aja yang paling mirip
local_alignment = pairwise2.align.localxx(dnaA, dnaB)
print("\nFormatted Local Alignment:")
for x in local_alignment:
    print(format_alignment(*x))
```

```python
# alignment dengan custom scoring

# Alignment global: bandingin dna secara keseluruhan
from Bio import pairwise2
global_alignment = pairwise2.align.globalms(dnaA, dnaB, 1, -1, -0.5, -0.1)  # match +2, mismatch -1, open gap -0.5, extend gap -0.1
print("Global Alignment:")
for x in global_alignment:
    print(x) # ada banyak alognment karena ada banyak cara align lalu diambil yg tertinggi

# print format
from Bio.pairwise2 import format_alignment
print("\nFormatted Global Alignment:")
for x in global_alignment:
    print(format_alignment(*x))


# Alignment local: bandingin dna secara lokal / pakai subsequence aja yang paling mirip
local_alignment = pairwise2.align.localms(dnaA, dnaB, 1, -1, -0.5, -0.1)
print("\nFormatted Local Alignment:")
for x in local_alignment:
    print(format_alignment(*x))
```

```python
# Distance: hamming: bikin function sendiri & levenshtein: pakai built in

# Hamming: hitung mismatch per posisi, harus sama panjang
def hamming(seq1:Seq, seq2:Seq) -> int:
    count = 0
    for x, y in zip (seq1, seq2):
        if x != y:
            count += 1
    return count

print("Hamming Distance:", hamming(dnaA, dnaB))


# Lavenshtein: hitung beda per edit (insert, delete, substitute), bisa beda panjang
%pip install python-Levenshtein

from Levenshtein import distance 
print("Levenshtein Distance:", distance(dnaA, dnaB))
```

```python
# Dot plot: visualisasi alignment
import matplotlib.pyplot as plt
import numpy as np

# cari yang match
def delta(a, b):
    return 1 if a != b else 0

def M(seq1, seq2, i, j, k):
    m = []
    for x, y in zip (seq1[i:i+k], seq2[j:j+k]):
        m.append(delta(x, y))
    return sum(m)

def makeMatrix(seq1, seq2, k):
    m1 = []
    for i in range(len(seq1)-k+1): # Biar ga out of bound
        m2 = []
        for j in range(len(seq2)-k+1):
            m2.append(M(seq1, seq2, i, j, k)) # k adalah ukuran window
        m1.append(m2)
    return m1

def plotMatrix(matrix, tolerance, seq1, seq2, nonblank='X', blank=' '):
    print(' | '+ seq2)
    print('-----------------')

    for label, row in zip(seq1, matrix):
        line = ''.join(nonblank if score < tolerance else blank for score in row)
        print(label+' | '+line)
        
def dotplot(seq1, seq2, k=1, tolerance=0):
    matrix = makeMatrix(seq1, seq2, k)
    plotMatrix(matrix, tolerance, seq1, seq2)
    
```
